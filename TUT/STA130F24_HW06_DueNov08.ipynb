4. Understand and explain the code below. Discuss with Chatbot how you could find the correlation coefficient through this code. Try to generate a similar code that find the correlation coefficient of your dataset from the first question of your homework.

```python
# Import necessary libraries
import pandas as pd
import statsmodels.formula.api as smf

# Example dataset
data = {
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Test_Score': [50, 55, 65, 70, 75]
}
df = pd.DataFrame(data)

# Calculate residuals and correlation coefficient
X = df['Hours_Studied']
y = df['Test_Score']

# Fit the model using statsmodels
model = smf.ols('Test_Score ~ Hours_Studied', data=df).fit()

# Get the residuals
residuals = model.resid
print("Residuals:", residuals)

# Get the correlation coefficient
correlation_coefficient = df.corr().loc['Hours_Studied', 'Test_Score']
print("Correlation Coefficient:", correlation_coefficient)

# Generate the summary table
summary = model.summary()
print(summary)
```    
<br><br>

5. Do you know what are the Simple Linear Regression (SLR) Principles? You recently received a dataset (under Fruther Guidance). This data shows the relationship between the amount of fertilizer used and crop yield. You would like to apply SLR, but your friend reminds you of the Simple Linear Regression Principles. They suggest you to double check the data you have, because it **doesn't** look right. Can you figure out which principles this data violates? 
<br><br>

<details class="details-example">
<summary style="color:blue"><u>Further Guidance</u></summary>
    
| Amount of Fertilizer (kg) (X) | Crop Yield (tons) (Y) |
|:-----------------------------:|:---------------------:|
| 1                             | 2                     |
| 2                             | 5                     |
| 3                             | 4                     |
| 4                             | 12                    |
| 5                             | 9                     |
| 6                             | 20                    |
| 7                             | 15                    |
| 8                             | 30                    |
| 9                             | 25                    |
| 10                            | 45                    |

     
    Hint: The Crop Yield seemed to be increasing too fast, that feels off... And the variance of residuals... huh, it doesn't feel right.
</details><br><br>

6. Recall the shuttlecocks example from your TUT demo. You are analyzing the relationship between Bird Flu Cases and Shuttlecock Price Increase using Simple Linear Regression (SLR). Try formulating the null and alternative hypotheses for testing the significance of the regression coefficient of Bird Flu Cases first, then identify the p-value for the regression coefficient of Bird Flu Cases using the summary function from smf.ols.fit(). <br><br> 

<details class="details-example">
    <summary style="color:blue"><u>Further Guidance</u></summary>
    <ul>
        <li> if you don't know how, Chatbot should be able to help you out. Remember to have 'price_increase ~ bird_flu_cases' as the formula.
        </li>
    </ul>
</details><br><br> 



7. You want to use bootstrapping to estimate the confidence interval for the correlation coefficient between Bird Flu Cases and Shuttlecock Price Increase and determine whether this interval contains zero. If zero is not in the interval, you would reject the null hypothesis. Ask Chatbot to help you understand the code here and how this code manage to find whether we should reject our null hypothesis at the 95% confidence level.

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

# Load your data
data = pd.DataFrame({
    'bird_flu_cases': [1112, 1985, 1142, 1528, 1741, 1002, 1266, 897, 1370, 1294, 892, 706, 1538, 1391, 1984, 1363, 1242, 1764, 1063, 1520],
    'price_increase': [5.01, 8.93, 4.44, 5.89, 7.31, 4.38, 5.22, 3.27, 6.10, 6.05, 3.59, 2.69, 7.38, 6.04, 8.77, 6.15, 5.08, 7.64, 4.95, 6.45]
})

# Fit the SLR model
model = smf.ols('price_increase ~ bird_flu_cases', data=data).fit()
print(model.summary())

# Function to calculate correlation coefficient
def correlation_coefficient(x, y):
    return np.corrcoef(x, y)[0, 1]

# Number of bootstrap samples
n_bootstrap = 1000
bootstrap_correlations = []

for _ in range(n_bootstrap):
    # Sample with replacement
    sample = data.sample(n=len(data), replace=True)
    corr = correlation_coefficient(sample['bird_flu_cases'], sample['price_increase'])
    bootstrap_correlations.append(corr)

# Calculate confidence intervals
lower_bound = np.percentile(bootstrap_correlations, 2.5)
upper_bound = np.percentile(bootstrap_correlations, 97.5)
print(f"95% Confidence Interval for the correlation coefficient: ({lower_bound}, {upper_bound})")

# Determine if we can reject the null hypothesis - What is the result?
...
```
